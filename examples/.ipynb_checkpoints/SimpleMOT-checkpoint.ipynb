```julia
using Revise
```

```julia
using OpticalBlochEquations
```

```julia
using Plots
using LinearAlgebra
using BenchmarkTools
using PhysicalConstants.CODATA2018
import DifferentialEquations: ODEProblem, solve, DP5, Tsit5, ExplicitRK, BS3, DP8
```

```julia
const h = PlanckConstant.val
const ħ = h / 2π
const μB = BohrMagneton.val
const c = SpeedOfLightInVacuum.val
const ϵ0 = VacuumElectricPermittivity.val

const λ = @with_unit 626 "nm"
const Γ = @with_unit 2π * 8.3 "MHz"
const M = @with_unit 50 "u"
const E₀ = c / λ
;
```

```julia
# Generate states
m1 = manifold(F=0, ω=0, μ=0)
m2 = manifold(F=1, ω=E₀/Γ, μ=0, Γ=Γ)
states = [m1.states..., m2.states...]
;
```

```julia
# Generate lasers
δ = -4Γ
s = 1.5
ẑ = [0, 0, 1]

l1 = define_laser(ẑ, ϵ₋, (E₀ + δ)/Γ, s)
l2 = define_laser(-ẑ, ϵ₋, (E₀ + δ)/Γ, s)

lasers = [l1, l2]
;
```

```julia
n = length(states)
d = zeros(n, n, 3)
d[1, 4, 3] = 1
d[1, 3, 2] = -1
d[1, 2, 1] = 1

for q in 1:3, i in 1:1, j in 2:4
    d[j,i,q] = d[i,j,q]
end
```

```julia
ρ0 = zeros(ComplexF64, 4, 4)
ρ0[1,1] = 1
(dρ, ρ, p) = obe(states, lasers, d, ρ0)
@btime ρ!($dρ, $ρ, $p, $1.0)
```

```julia
tspan = (0., 100); tstops = 0:0.1:600
prob = ODEProblem(ρ!, ρ, tspan, p)
@time sol = solve(prob, alg=DP5(), abstol=1e-6, reltol=1e-3, dense=false)
;
```

```julia
length(sol.t)
```

```julia
ρ₁₁ = [real(x[1,1]) for x in sol.u]
ρ₂₂ = [real(x[2,2]) for x in sol.u]
ρ₃₃ = [real(x[3,3]) for x in sol.u]
ρ₄₄ = [real(x[4,4]) for x in sol.u]
plot(sol.t, ρ₁₁)
plot!(sol.t, ρ₂₂)
plot!(sol.t, ρ₃₃)
plot!(sol.t, ρ₄₄)
```
