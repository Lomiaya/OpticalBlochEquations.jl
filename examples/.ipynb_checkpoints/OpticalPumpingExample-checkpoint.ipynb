```julia
using Revise
```

```julia
using OpticalBlochEquations
```

```julia
using Plots
using LinearAlgebra
using BenchmarkTools
using PhysicalConstants.CODATA2018
import DifferentialEquations: ODEProblem, solve, DP5, Tsit5, ExplicitRK, BS3, DP8
```

```julia
const μB = BohrMagneton.val
const ϵ0 = VacuumElectricPermittivity.val

const λ = @with_unit 626 "nm"
const Γ = @with_unit 2π * 8.3 "MHz"
const M = @with_unit 57 "u"
const E₀ = c / λ
;
```

```julia
# Generate states
m1 = manifold(F=2, ω=0, μ=0)
m2 = manifold(F=3, ω=E₀/Γ, μ=0, Γ=Γ)
states = [m1.states..., m2.states...]
;
```

```julia
# Generate lasers
δ = -2.73Γ
x̂ = [1,0,0]
ŷ = [0,1,0]
pol = [0,1,0]

laser1 = Laser(x̂, pol, (E₀ + δ)/Γ, 4*1.352464)
laser2 = Laser(-x̂, pol, (E₀ + δ)/Γ, 4*1.352464)
# laser = define_laser(x̂, [0,0,1], (E₀ + δ)/Γ, 4*1.352464)
lasers = [laser2]
;
```

```julia
d = zeros(12, 12, 3)
d[1, 8, 1] = 0.25819889
d[2, 9, 1] = 0.4472136
d[3, 10, 1] = 0.63245553
d[4, 11, 1] = 0.81649658
d[5, 12, 1] = 1

d[1, 7, 2] = -0.57735027
d[2, 8, 2] = -0.73029674
d[3, 9, 2] = -0.77459667
d[4, 10, 2] = -0.73029674
d[5, 11, 2] = -0.57735027

d[1, 6, 3] = 1
d[2, 7, 3] = 0.81649658
d[3, 8, 3] = 0.63245553
d[4, 9, 3] = 0.4472136
d[5, 10, 3]= 0.25819889

for q in 1:3, i in 1:5, j in 6:12
    d[j,i,q] = d[i,j,q]
end
```

```julia
ρ0 = zeros(ComplexF64, length(states), length(states))
ρ0[1,1] = 1

(dρ, ρ, p) = obe(states, lasers, d, ρ0)
@btime ρ!($dρ, $ρ, $p, $1.0)
```

```julia
saved_values = SavedValues(Float64, Float64)
cb = SavingCallback((u, t, integrator) -> force(integrator.p), saved_values)
;
```

```julia
tspan = (0., 600); tstops = 0:0.1:600
prob = ODEProblem(ρ!, ρ, tspan, p, callback=AutoAbstol(false, init_curmax=0.0))
@time sol = solve(prob, alg=DP5(), abstol=1e-6, reltol=1e-3, dense=false)
length(sol.t)
```
